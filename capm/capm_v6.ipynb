# Resampling as monthly data

```python
df_monthly = df.resample('M', convention='end').sum()
```

# Computing the Beta

```python
df_2016 = df_monthly['2016-01-01':'2020-05-31']
```

```python
betas = []
for ticker in df_2016.columns:
    for index in df_2016.index:
         betas.append(np.cov(df_2016[ticker][:index],df_2016['Index'][:index])[0][1] / np.cov(df_2016[ticker][:index],df_2016['Index'][:index])[1][1])
```

```python
betas_array = np.array_split(np.array(betas), df_2016.columns.size)
```

```python
betas_df = pd.DataFrame()
for stock in betas_array:
    betas_df = pd.concat([betas_df, pd.Series(stock)], axis=1)
```

```python
betas_df.index = df_2016.index
betas_df.columns = df_2016.columns
```
