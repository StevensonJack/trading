# II - Indices

```python
data_tickers = pd.read_csv("data/FSE_metadata.csv")
```

```python
tickers = np.array(data_tickers["code"])
```

```python
tickers
```

```python
FSE = []
start = time.time()
for ticker in tickers:
    FSE.append(quandl.get(f"FSE/{ticker}.4"))
end = time.time()
print(end - start)
```

```python
FSE
```

```python
df = pd.DataFrame()
for stock in FSE:
    df = pd.concat([df, stock], axis=1)
```
